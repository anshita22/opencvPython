```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('pooh.jpg')
plt.imshow(image)
plt.show()
img= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
new_image= cv2.resize(img, (100,100), interpolation= cv2.INTER_LINEAR)           #Bilinear interpolation
plt.imshow(new_image)
plt.show()
```

```python
new_image= cv2.resize(img, (100,100), interpolation= cv2.INTER_AREA)   #Resampling using pixel area relation
plt.imshow(new_image)
plt.show()
```

```python
new_image= cv2.resize(img, (100,100), interpolation= cv2.INTER_CUBIC)   #Bicubic interpolation over 44 pixel neighborhood
plt.imshow(new_image)
plt.show()
```

```python
new_image= cv2.resize(img, (100,100), interpolation= cv2.INTER_NEAREST)   #Nearest neighbor interpolation
plt.imshow(new_image)
plt.show()
```

```python
new_image= cv2.resize(img, (100,100), interpolation= cv2.INTER_LANCZOS4)   #Lanczos interpolation over 88 neighborhood
plt.imshow(new_image)
plt.show()
```
