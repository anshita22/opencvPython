```python
import cv2
import matplotlib.pyplot as plt
```

```python
"""detecting faces using haarcascade classifier, A haarcascade is basically  a classifier which is used to detect particular 
object from the source"""
```

```python
#haarcascade_frontalface_default.xml is a haarcascade designed by opencv to detect frontal face 
detect = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
imp_img= cv2.VideoCapture("elon.jpg")
res, img= imp_img.read()
gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces= detect.detectMultiScale(gray, 1.3, 5)
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 255, 0), 3)
img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```
